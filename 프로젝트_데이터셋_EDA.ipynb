# 1.1 Data Dictionary

## 1. age: age in years

## 2. sex: sex
- 1 = male
- 0 = female

## 3. cd : condition type
- Value 0: Internal Medicine
- Value 1: General Surgery
- Value 2: Chest surgery
- Value 3: Urology
- Value 4: Ear & Nose & Throat, otolaryngology

## 4. Procedure type
- Value 0: surgery
- Value 1: medication
- Value 2: various tests
- Value 3: Consultation and Management

## 5. Cost
- 0 = Cost > 8,370$
- 1 = Cost < 8,370$

## 6. Length_of_Stay
- Length_of_Stay > 37 = 1
- Length_of_Stay < 37 = 0

## 7. Readmission
- 0 = YES
- 1 = NO

## 8. Satisfaction
- 1 = Satisfaction < 3
- 0 = Satisfaction > 3

## 9. target : Outcome
- 0 = Recovered
- 1 = Stable
